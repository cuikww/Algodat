```python
from pyspark import SparkContext
```

```python
sc = SparkContext.getOrCreate()
```

```python
data = [i for i in range(10000)]
```

```python
rdd = sc.parallelize(data)
```

```python
rdd.count()
```

```python
len(data)
```

```python
rdd_map = rdd.map(lambda x:x*2)
```

```python
rdd_map.collect()
```

```python
def kalidua(x):
  if x%4 == 0:
    return x*4
  else:
    return x*10
```

```python
rdd_map = rdd.map(kalidua)
rdd_map.collect()
```

```python
rdd_reduce = rdd.reduce(lambda a,b:a+b)
rdd_reduce
```

```python
rdd_reduce = rdd_map.reduce(lambda a,b:a+b)
rdd_reduce
```

```python
def total(a,b):
  if(b-a)%2==0:
    return 3*a+2*b
  else:
    return 2*a+3*b
```

```python
rdd_reduce = rdd.reduce(total)
rdd_reduce
```

```python
rdd_filter = rdd.filter(lambda x: x%105==0)
rdd_filter.collect()
```
